```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(image):
    lower_bound = np.array([5,10,15])
    upper_bound = np.array([10,150,200])
    mask = cv2.inRange(image, lower_bound, upper_bound)
    img = cv2.bitwise_and(image, image, mask=mask)
    return img

im0 = cv2.imread('yo.jpg', cv2.IMREAD_COLOR)
im = func(im0)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.xlabel('Original Image')
plt.imshow(im0)
plt.subplot(1,2,2)
plt.xlabel('Final Image')
plt.imshow(im)
```

<b>Why Simple Grayscale May Not Always Work</b> <br>
-Loss of Detail: Important color features are lost, making it harder to identify objects. <br>
-Ambiguity: Similar grayscale values can represent different colors, leading to confusion. <br>
-Selective Analysis: Grayscale doesnt allow for isolating specific colors for analysis. <br><br>
